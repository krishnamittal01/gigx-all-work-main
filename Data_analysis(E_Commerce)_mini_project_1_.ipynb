**E-COMMERCE ANALYSIS**

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
import plotly.colors as colors
import plotly.io as pio
pio.templates.default='plotly_white'
```

```python
df=pd.read_csv("/content/Sample - Superstore.csv",encoding='latin-1')  ## This encoding=latin-1 is used to read special characters
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
### order date and ship date are in string but we wanna change it to date type
```

**Cnverting Date columns**

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Ship Date'])
```

```python
df.info()
```

**Monthly Sales Analysis**

```python
## make 3 diff columns of month, order day of week , year
```

```python
df['Order Month']=df['Order Date'].dt.month
df["Order Year"]=df['Order Date'].dt.year
df['Oredr Day of Week']=df['Order Date'].dt.dayofweek
```

```python
df.info()
```

```python
## Now finally analyse the monthly sales
```

```python
sales_by_month=df.groupby("Order Month")['Sales'].sum().reset_index()
```

```python
sales_by_month.head()
```

```python
fig=px.line(sales_by_month,x='Order Month',y='Sales',title='Monthly Sales Analysis')
fig
```

**Sale Based on Categories**

```python
sales_by_categories=df.groupby('Category')['Sales'].sum().reset_index()
```

```python
fig_2=px.pie(sales_by_categories,values='Sales',names='Category',title='Sale Based on Categories',hole=0.4,color_discrete_sequence=px.colors.qualitative.Pastel)
fig_2.update_traces(textposition='outside',textinfo='percent+label')
fig_2.update_layout(title_font=dict(size=24))
fig_2
```

**Sales Analysis by Sub Category**

```python
sales_by_subcategory=df.groupby('Sub-Category')['Sales'].sum().reset_index()
```

```python
fig_3=px.bar(sales_by_subcategory,x='Sub-Category',y='Sales',title="Sales Analysis By Sub Category")
fig_3
```

**Analyze The Monthly Profit**

```python
profit_by_Month=df.groupby("Order Month")['Profit'].sum().reset_index()
```

```python
fig_4=px.bar(profit_by_Month,x='Order Month',y='Profit',title='Monthly Profit Analysis')
fig_4
```

**Profit By Category**

```python
profit_by_category=df.groupby('Category')['Profit'].sum().reset_index()
```

```python
fig_5=px.pie(profit_by_category,values='Profit',names='Category',title='Profit By Category',hole=0.4,color_discrete_sequence=px.colors.qualitative.Pastel)
fig_5.update_traces(textposition='outside',textinfo='percent+label')
fig_5.update_layout(title_font=dict(size=30))

fig_5
```

**Profit By Sub-Category**

```python
profit_by_Subcategory=df.groupby('Sub-Category')['Profit'].sum().reset_index()
fig_6=px.line(profit_by_Subcategory,x='Sub-Category',y='Profit',title='Profit By Sub-Category')
fig_6
```

**Sales And Profit According To Customer Segment**

```python
sales_profit_by_segment=df.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
fig_7=go.Figure()
fig_7.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                       y=sales_profit_by_segment['Sales'],
                       name='Sales',marker_color='blue'))
fig_7.add_trace(go.Bar(x=sales_profit_by_segment['Segment'],
                       y=sales_profit_by_segment['Profit'],
                       name='Profit',marker_color='green'))
fig_7.update_layout(title='Sales And Profit According To Customer Segment',
                    xaxis_title='Customer Segment',
                    yaxis_title='Amount',
                    barmode='group')
fig_7
```

**Sales To Profit Ratio**

```python
sales_profit_by_segment=df.groupby('Segment').agg({'Sales':'sum','Profit':'sum'}).reset_index()
sales_profit_by_segment['Sales To Profit Ratio']=sales_profit_by_segment['Sales']/sales_profit_by_segment['Profit']
print(sales_profit_by_segment[['Segment','Sales To Profit Ratio']])
```
